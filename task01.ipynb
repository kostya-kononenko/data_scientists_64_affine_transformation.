```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
chess_1, chess_2 = cv2.imread('chess_1.png'), cv2.imread('chess_2.png')
```

```python
fig, axis = plt.subplots(1, 2)
axis[0].imshow(chess_1)
axis[1].imshow(chess_2)
```

```python
chess_1_points = np.array([[196, 53], [445, 43], [201, 243], [440, 256]])
chess_2_points = np.array([[192, 316], [173, 62], [406, 362], [514, 127]])
```

```python
h, status = cv2.findHomography(chess_2_points, chess_1_points)
```

```python
warp_img = cv2.warpPerspective(chess_2, h, (chess_2.shape[1], chess_2.shape[0]))
```

```python
plt.imshow(warp_img)
```
